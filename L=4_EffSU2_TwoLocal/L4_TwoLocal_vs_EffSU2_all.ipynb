# Lambda = 4: TwoLocal vs EfficientSU2

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
su2_ansatz = ['effsu2_Ry_c', 'effsu2_Rz_c', 'effsu2_RyRz_c', 'effsu2_RyY_c',
             'effsu2_Ry_f', 'effsu2_Rz_f', 'effsu2_RyRz_f', 'effsu2_RyY_f']
tl_ansatz = ['tl_Ry_c', 'tl_Rz_c', 'tl_RyRz_c', 'tl_RyY_c',
             'tl_Ry_f', 'tl_Rz_f', 'tl_RyRz_f', 'tl_RyY_f']
```

```python
def plot_tl_vs_ef(df1, df2, ll1, ll2, E, opt, Lambda):
    L = len(df1.columns)
    fig,ax = plt.subplots(int(L/2),2, figsize = (10,15), tight_layout = True)
    for i in range(0,L,2):
        ax[int(i/2),0].plot(np.arange(len(df1[df1.columns[i]])), df1[df1.columns[i]], label = df1.columns[i])
        ax[int(i/2),0].plot(np.arange(len(df2[df2.columns[i]])), df2[df2.columns[i]], label = df2.columns[i])
        ax[int(i/2),0].axhline(E, color = 'black', linestyle='dashed', label='Exact energy')
        ax[int(i/2),0].legend()
    
        ax[int(i/2),1].plot(np.arange(len(df1[df1.columns[i+1]])), df1[df1.columns[i+1]], label = df1.columns[i+1])
        ax[int(i/2),1].plot(np.arange(len(df2[df2.columns[i+1]])), df2[df2.columns[i+1]], label = df2.columns[i+1])
        ax[int(i/2),1].axhline(E, color =  'black', linestyle='dashed', label='Exact energy')
        ax[int(i/2),1].legend()
    plt.suptitle(f'$\Lambda = {Lambda}, \lambda = {ll1}$: TwoLocal versus EfficientSU2 (with {opt})', fontsize = 18)
    fname = f'results/l={ll2}_tl_vs_effsu2_{opt}'
    plt.savefig(fname)
    print(f'Figure saved as {fname}')
```

# lambda = 0.2

```python
E_02 = 3.13406
df_02_cs= pd.read_csv('results/qve_l4_l=0.2_effsu2_cobyla.csv')
df_02_ct= pd.read_csv('results/qve_l4_l=0.2_twolocal_cobyla.csv')
df_02_ss= pd.read_csv('results/qve_l4_l=0.2_effsu2_spsa.csv')
df_02_st= pd.read_csv('results/qve_l4_l=0.2_twolocal_spsa.csv')
```

```python
df_02_cs.drop(['Unnamed: 0'], axis = 1, inplace=True)
df_02_cs.columns = su2_ansatz

df_02_ct.drop(['Unnamed: 0'], axis = 1, inplace=True)
df_02_ct.columns = tl_ansatz

df_02_ss.drop(['Unnamed: 0'], axis = 1, inplace=True)
df_02_ss.columns = su2_ansatz

df_02_st.drop(['Unnamed: 0'], axis = 1, inplace=True)
df_02_st.columns = tl_ansatz
```

```python
plot_tl_vs_ef(df_02_ss, df_02_st,'0.2','02', E_02, 'SPSA', 4)
```

```python
plot_tl_vs_ef(df_02_cs, df_02_ct,'0.2','02', E_02, 'COBYLA', 4)
```
